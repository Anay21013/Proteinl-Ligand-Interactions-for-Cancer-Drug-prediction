```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
combined_df = pd.read_csv('./combined_proteins.csv')
```

```python
combined_df
```

```python
combined_df['ic50'] = pd.to_numeric(combined_df['ic50'], errors='coerce')
```

```python
combined_df = combined_df.dropna(subset=['ic50'])
combined_df['ic50']
```

```python
combined_df['ic50'] = combined_df['ic50'].astype(int)
```

```python
cnt=0
for i in range(len(combined_df)):
    if (combined_df['ic50']).tolist()[i] > 0:
        print((combined_df['ic50']).tolist()[i])
        cnt += 1
cnt
```

```python
ligand_features = combined_df.columns[47:1873].tolist()
```

```python
len(ligand_features)
```

```python
protein_features = combined_df.columns[1873:2834].tolist()
```

```python
len(protein_features)
```

```python
combined_df[ligand_features]
```

```python
combined_df[protein_features]
```

```python
combined_df[ligand_features] = combined_df[ligand_features].apply(pd.to_numeric, errors='coerce')
combined_df[protein_features] = combined_df[protein_features].apply(pd.to_numeric, errors='coerce')
```

```python
combined_df[ligand_features] = combined_df[ligand_features].replace([np.inf, -np.inf], np.nan).fillna(0)
combined_df[protein_features] =combined_df[protein_features].replace([np.inf, -np.inf], np.nan).fillna(0)
```

```python
combined_df[ligand_features] = combined_df[ligand_features].round(decimals=6)
combined_df[protein_features] = combined_df[protein_features].round(decimals=6)
```

```python
combined_df
```

```python
# Features and target variable
X_ligand = combined_df[ligand_features]
X_protein = combined_df[protein_features]
y = combined_df['ic50']
```

```python
X = pd.concat([X_ligand,X_protein],axis=1)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Predict IC50 values on the test set
y_pred = model.predict(X_test)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Calculate R-squared (R²)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R²): {r2}")
```
