```python
import pandas as pd
from rdkit import Chem
from rdkit.ML.Descriptors import MoleculeDescriptors
from rdkit.Chem import Descriptors
```

```python
descriptor_names = [desc[0] for desc in Descriptors._descList]
descriptor_names
```

```python
calculator = MoleculeDescriptors.MolecularDescriptorCalculator(descriptor_names)
calculator
```

```python
def calculate_descriptors(smiles):
    if not isinstance(smiles, str) or not smiles:
        # Return None for each descriptor if the SMILES string is invalid
        return [None] * len(descriptor_names)
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        return calculator.CalcDescriptors(mol)
    else:
        return [None] * len(descriptor_names)
```

```python
# Calculate descriptors for each SMILES string and create a DataFrame
descriptor_values = df['canonical_smiles'].apply(calculate_descriptors)
descriptor_df = pd.DataFrame(descriptor_values.tolist(), columns=descriptor_names)
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
assert len(df) == len(descriptor_df)
```

```python
def generate_3d_structure(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return None
        mol = Chem.AddHs(mol)  # Add hydrogens
        AllChem.EmbedMolecule(mol, AllChem.ETKDG())  # Generate 3D coordinates
        AllChem.UFFOptimizeMolecule(mol)  # Optimize the 3D structure
        # print("done")
        return mol
    except Exception as e:
        print(f"Failed to generate 3D structure for SMILES: {smiles}, error: {e}")
        return None
```

```python
# Apply the function to the original DataFrame 'df' to generate 3D structures
df['mol_3d'] = df['canonical_smiles'].apply(generate_3d_structure)
```

```python
df = df.dropna(subset=['mol_3d'])
```

```python
import pandas as pd
import requests
from protlearn.features import aaindex1, geary, ngram
from rdkit import Chem
from rdkit.Chem import AllChem
from mordred import Calculator, descriptors
```

```python
# Function to fetch the ALK protein sequence from UniProt
def fetch_uniprot_sequence(uniprot_id):
    url = f"https://www.uniprot.org/uniprot/{uniprot_id}.fasta"
    response = requests.get(url)
    if response.status_code == 200:
        fasta_data = response.text
        sequence = ''.join(fasta_data.split('\n')[1:])
        return sequence
    else:
        raise ValueError(f"Failed to fetch sequence for UniProt ID {uniprot_id}")
```

```python
# UniProt ID for BRAF protein
uniprot_id = 'P15056'
braf_sequence = fetch_uniprot_sequence(uniprot_id)
```

```python
aaindex_features, aaindex_labels = aaindex1(braf_sequence)
```

```python
aaindex_features
```

```python
geary_features = geary(braf_sequence)
geary_labels = [f'geary_{i+1}' for i in range(len(geary_features[0]))]
```

```python
ngram_features, ngram_labels = ngram(braf_sequence, n=2)
```

```python
# Combine all features into a single DataFrame with labels
braf_descriptors = pd.concat([
    pd.DataFrame([aaindex_features][0], columns=aaindex_labels),
    pd.DataFrame([geary_features][0], columns=geary_labels),
    pd.DataFrame([ngram_features][0], columns=ngram_labels)
], axis=1)
```

```python
braf_descriptors = braf_descriptors.iloc[0].to_frame().T
```

```python
braf_descriptors
```

```python
for col in braf_descriptors.columns:
    combined_df[col] = braf_descriptors[col].values[0]
```

```python
combined_df['ic50'] = 10**(9 - combined_df['standard_value'])
```
