```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import PandasTools
from chempy import Substance
```

```python
alk_path = './01_ChEBML_ALK.csv'
df = pd.read_csv(alk_path)
```

```python
def generate_3d_structure(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None
    mol = Chem.AddHs(mol)
    AllChem.EmbedMolecule(mol, AllChem.ETKDG())
    AllChem.UFFOptimizeMolecule(mol)
    return mol

df['3d_structure'] = df['canonical_smiles'].apply(generate_3d_structure)
```

```python
from padelpy import from_smiles
import subprocess
```

```python
def calculate_descriptors(smiles):
    try:
        descriptors = from_smiles(smiles)
        return descriptors
    except Exception as e:
        print(f"Error calculating descriptors for {smiles}: {e}")
        return {}
```

```python
descriptors_list = df['canonical_smiles'].apply(calculate_descriptors)
```

```python
descriptors_df = pd.json_normalize(descriptors_list)
```

```python
result_df = pd.concat([df, descriptors_df], axis=1)
```

```python
import pandas as pd
import requests
from protlearn.features import aaindex1, geary, ngram
from rdkit import Chem
from rdkit.Chem import AllChem
from mordred import Calculator, descriptors
```

```python
# Function to fetch the ALK protein sequence from UniProt
def fetch_uniprot_sequence(uniprot_id):
    url = f"https://www.uniprot.org/uniprot/{uniprot_id}.fasta"
    response = requests.get(url)
    if response.status_code == 200:
        fasta_data = response.text
        sequence = ''.join(fasta_data.split('\n')[1:])
        return sequence
    else:
        raise ValueError(f"Failed to fetch sequence for UniProt ID {uniprot_id}")
```

```python
# UniProt ID for ALK protein
uniprot_id = 'Q9UM73'
alk_sequence = fetch_uniprot_sequence(uniprot_id)
```

```python
aaindex_features, aaindex_labels = aaindex1(alk_sequence)
```

```python
geary_features = geary(alk_sequence)
geary_labels = [f'geary_{i+1}' for i in range(len(geary_features[0]))]
```

```python
ngram_features, ngram_labels = ngram(alk_sequence, n=2)
```

```python
# Combine all features into a single DataFrame with labels
alk_descriptors = pd.concat([
    pd.DataFrame([aaindex_features][0], columns=aaindex_labels),
    pd.DataFrame([geary_features][0], columns=geary_labels),
    pd.DataFrame([ngram_features][0], columns=ngram_labels)
], axis=1)
```

```python
alk_descriptors = alk_descriptors.iloc[0].to_frame().T
```

```python
alk_descriptors
```

```python
for col in alk_descriptors.columns:
    result_df[col] = alk_descriptors[col].values[0]
```
